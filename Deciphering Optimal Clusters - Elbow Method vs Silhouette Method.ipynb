# Deciphering Optimal Clusters: Elbow Method vs. Silhouette Method
Blog with detailed explanation can be found [here](https://medium.com/p/7e311c604201/)


```python
from sklearn.datasets import make_blobs
import numpy as np

# Generate synthetic data
# We are creating 150 data points, grouped into 3 clusters
X, _ = make_blobs(n_samples=150, centers=3, cluster_std=0.60, random_state=0)
# If you want to visualize these clusters, you can use the following code
import matplotlib.pyplot as plt
plt.scatter(X[:,0], X[:,1], s=50)
plt.title('Synthetic Data Points')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
range_values = range(1, 10)

for i in range_values:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.plot(range_values, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

silhouette_scores = [] 
range_values = range(2, 10)

for i in range_values:
    kmeans = KMeans(n_clusters=i).fit(X)
    labels = kmeans.labels_
    silhouette_scores.append(silhouette_score(X, labels, metric = 'euclidean'))

plt.plot(range_values, silhouette_scores)
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Method showing the optimal k')
plt.show()
```
